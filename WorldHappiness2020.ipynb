```python
import numpy as np
import seaborn as sn
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as po
import plotly.graph_objs as pg
from plotly.offline import iplot
import os
```

## **Data processing**


```python
df = pd.read_csv("https://raw.githubusercontent.com/PriyankaDhar2/PriyankaDhar2/main/2020.csv")
df.head()
```

```python
df.dropna()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(columns=['Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Ladder score in Dystopia'])
df.head()
```

```python
df.shape
```

```python
col_names_dict = {'Country name':'Country', 'Regional indicator':'Region',
                  'Logged GDP per capita': 'GDPPC','Ladder score':'Happiness score',
                  'Social support':'Social Support', 'Healthy life expectancy':'Life Expectancy',
                  'Freedom to make life choices':'Freedom', 'Perceptions of corruption': 'Corruption'}

df.rename(columns = col_names_dict, inplace = True)
df.head()
```

```python
fig = px.scatter(df,x = 'Happiness score', y = 'GDPPC', color = 'Region',)
fig.show()
```

## visualizing the happiness scores into an interactive world map

```python
df.columns
```

Maps

```python
map_happy = dict(type = 'choropleth', 
           locations = df['Country'],
           locationmode = 'country names',
           z = df['Happiness score'], 
           text = df['Country'],
           colorbar = {'title':'Happiness score scale'},
           colorscale = 'curl')
layout = dict(title = 'Happiness score of different countries across the world', 
              geo = dict(showframe = False, projection = {'type': 'equirectangular'}))

choromap3 = go.Figure(data = [map_happy], layout=layout)
iplot(choromap3)
```

```python
data = dict(type='choropleth', 
            locations = df['Country'],
           locationmode = 'country names', 
            colorscale = 'portland',
            z = df['GDPPC'], 
            text = df['Country'])
```

```python
layout = dict(title = 'Global GDP',
              geo = dict( projection = {'type':'hammer'}))
x = pg.Figure(data = [data], 
              layout = layout)
po.iplot(x)
```
